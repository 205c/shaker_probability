Yossi's Question:
-----------------
```
"My question is this, say I’m running the following A/B test:
50% of my users see variation A of a “buy dialog", 50% variation B of a “buy dialog".
I’m interested in finding the dialog that produces the most purchases.

How do I know in advance the sample size needed to answer this question for a decent level of certainty?
Say the experiment is already running, how do I know if I can go ahead and stop it because a certain variation won?"
```

````python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
````

Preliminaries
-------------

````python
X = np.linspace(0,1)[1:-1] #the different values that X can take on (0 to 1)
Y = np.linspace(0,1)[1:-1] #the different values that Y can take on (0 to 1)
a = (np.random.randn(10) > 0) #10 bernoulli trials for product A, recorded 
b = (np.random.randn(10) > 0) #10 bernoulli trials for product B, recorded
````

````python
num_success = sum(a)
num_failure = len(a) - sum(a)

p_X_given_a = (X**num_success)*((1-X)**(num_failure))
p_X_given_a /= p_X_given_a.sum() #normalize to get prob dist.

plt.plot(X, p_X_given_a)
plt.xlabel("possible values of X")
plt.ylabel("proportional probability of X given values")
plt.title("P(X|a_1...a_n)")
````

````python
p_Y_given_b = (Y**sum(b))*((1-Y)**(len(b)-sum(b)))
p_Y_given_b /= p_Y_given_b.sum() #normalize to get probability distribution.
````

````python
joint_prob = np.outer(p_X_given_a, p_Y_given_b)

plt.imshow(joint_prob)
plt.xlabel("values of X (0-1)")
plt.ylabel("values of Y (0-1), top=0")
# plt.gca().set_xticklabels(np.linspace(0,1)[1:-1])
# plt.gca().set_yticklabels(Y.round(2))
plt.gca().set_xticklabels([])
plt.gca().set_yticklabels([])
plt.title("P(X,Y|a_1...a_n, b_1...b_n)")
plt.show ()
````

````python
#Note that the region where X > Y is precisely the region beneath the diagonal.
region_of_interest = np.triu(joint_prob)

plt.imshow(region_of_interest)
plt.title("the region we will integrate over")
plt.xlabel("Values of X (0-1)")
plt.ylabel("Values of Y (0-1, top=0)")
plt.gca().set_xticklabels([])
plt.gca().set_yticklabels([])
````

````python
print "Probability that X > Y is currently %s" % np.sum(np.sum(region_of_interest))
````

So, how to choose the sample size?
----------------------------------

There's a pretty good answer here that utilizes the Chernoff Bound (or Hoeffding Inequality, depending on how you learned it). Say you have n samples from a Bernoulli distribution. Then the Hoeffding inequality gives us a bound x, such that the chance the sample mean is within a given distance from the true mean is greater than or equal to x. 

So, let's apply it to the case of your A/B testing. The test you described is a Bernoulli trial; each person either accepts or declines the dialog. We're interested in finding the true means of the two Bernoulli distributions, one for A, one for B; this is the percentage of people that accept the dialog. If A's true mean is greater than B's true mean, A is better; else B is better.

Hoeffding's inequality gives us a lower bound on the probability that the true mean is within a certain distance gamma of the sample mean, as a function of the sample size. So if we solve for the sample size and supply a lower bound on certainty and a certain distance gamma, we'll get the answer to your question.

First pick a certainty level; if you want 95% certainty, set the lower bound to 0.95; then with probability at least 0.95, the true mean is within the certain distance gamma of the sample mean. 

Gamma is how close we want to get to the true mean. For a good choice of gamma, note that a problem we might run into using Hoeffding bounds is if the two gamma intervals overlap: perhaps A's sample mean is below B's sample mean, but within gamma of each other, so that we can't be sure A's true mean is below B's true mean. 

To get around this, first decide how close the two true means need to be for you to say that A and B are "practically the same". 0.01 might be fair; that says that if the proportion of people in A that accept the dialog is < 1% away from the proportion of people in B that accept the dialog, the two dialogs are practically the same. Then choose gamma = HALF of that number. Then if the intervals overlap at all, we know that the true means are within 1% of each other, and we can conclude that A and B are practically the same.
